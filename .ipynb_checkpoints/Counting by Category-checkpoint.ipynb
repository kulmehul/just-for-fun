```python
# Given a list of items, which category has most items?
```

```python
import re
```

```python
list_of_items = ["Star",1,"A","B",2,"Star","Table"]
```

```python
# Pehle to mujhe inhe category me daalna padega, fir apan inhe ginenge...
# Maan lete h ya to mere paas number h, ya mere paas alphabet h ya mere paas object h..
objects = []
alphabets = []
numbers = []

for item in list_of_items:
    x = re.search("\d", str(item))
    y = re.findall("[a-zA-Z]",str(item))
    # Sir aapko kaise samjhe? Aapka kaam kya h? Dekho m dhundhta hu.
    # Agar mujhe kuch milta h, to m kuch deta hu, nhi to m kuch nhi deta. Yani None deta hu.
    # Matlab, None means sorry.
    # Means yha x ko padhna h, if x m kuch h... 
    if(x):
        numbers.append(item)        # Append bole to isme daalo
    elif(len(y)==1):
        alphabets.append(item)
    else:
        objects.append(item)
    
```

```python
kitne_item_in_category = {"numbers":len(numbers), 
                          "alphabets": len(alphabets),
                          "objects": len(objects)}
```

```python
print(kitne_item_in_category)
```

```python
# Ye wala logic category k numbero ko dekh k bta dega ki ye wala highest h
chiz_jiski_max_value_h = max(zip(kitne_item_in_category.values(),kitne_item_in_category.keys()))
```

```python
print("The number of",chiz_jiski_max_value_h[1],"is sabse zyada. ",chiz_jiski_max_value_h[0],"count h iska yr..")
```
